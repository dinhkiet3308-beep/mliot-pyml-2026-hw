### MEAN, MEDIAN, MODE, VAR, STD, MIN, MAX, Q123, IQR

Pandas lib has a built-in function that can calculate most of these for us.

```python
print(Iris_sns.describe())
```

Or we can calculate by each column.

```python
mean1 = Iris_sns["sepal_length"].mean()
print(f"mean of sepal_length: {mean1}")

mean2 = Iris_sns["sepal_width"].mean()
print(f"mean of sepal_width: {mean2}")

mean3 = Iris_sns["petal_length"].mean()
print(f"mean of petal_length: {mean3}")

mean4 = Iris_sns["petal_width"].mean()
print(f"mean of petal_width: {mean4}")

# or we can calculate mean by formula
def mean(query):
    total = 0
    for i in range(len(Iris_sns[query])):
        total += Iris_sns[query][i]
    return (f"mean of {query} is: {total/len(Iris_sns[query])}")

mean("sepal_length")
```

```python
median1 = Iris_sns["sepal_length"].median()
print(f"median of sepal_length: {median1}")

median2 = Iris_sns["sepal_width"].median()
print(f"median of sepal_width: {median2}")

median3 = Iris_sns["petal_length"].median()
print(f"median of petal_length: {median3}")

median4 = Iris_sns["petal_width"].median()
print(f"median of petal_width: {median4}")

#or we can calculate median by formula
def median(query):
    if len(Iris_sns[query]) % 2 == 0:
        pos = len(Iris_sns[query]) / 2 - 1
        return (f"median of {query}")
```

Note: median = Q2, lies at the middle of the data (50%).

```python
mode1 = Iris_sns["sepal_length"].mode()[0]
print(f"mode of sepal_length: {mode1}")

mode2 = Iris_sns["sepal_width"].mode()[0]
print(f"mode of sepal_width: {mode2}")

mode3 = Iris_sns["petal_length"].mode()[0]
print(f"mode of petal_length: {mode3}")

mode4 = Iris_sns["petal_width"].mode()[0]
print(f"mode of petal_width: {mode4}")

mode = Iris_sns["sepal_length"].mode()
print(mode)
```

mode() outputs a pandas array (dictionary) has matches the actual mode to an index. In this case is 0 as there's only one mode.  
But if the data has more than 2 modes the index will continue 0 -> 1 -> 2, so we have to call mode()[0] not mode()[1]
